```python
import gensim.downloader
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
import re
import seaborn as sns

from nltk.tokenize import word_tokenize
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tags = [
    'javascript',
    'python',
    'java',
    'c_sharp',
    'php',
    'android',
    'html',
    'jquery',
    'c_plus_plus',
    'css',
    'ios',
    'sql',
    'mysql',
    'r',
    'reactjs',
    'node_js',
    'arrays',
    'c',
    'asp_net',
    'json'
]
```

```python
# I obtained the data by running this query in the Stack Exchange Data Explorer.
# This consists in questions having all at least 50 likes, 5 answers and 1 tag
# among the 20 most common tags.
# For modeling purposes, I kept 500 questions per tag: the questions with the
# highest number of likes.
# Some questions have several tags among the 20 most common tags.
# So the total number of questions is less than 500 * 20 = 10,000.

sql = """
WITH Tags AS
(
	SELECT '<javascript>' 	AS tag_string, 'javascript' 	AS tag_name UNION ALL
	SELECT '<python>' 		AS tag_string, 'python' 		AS tag_name UNION ALL
	SELECT '<java>' 		AS tag_string, 'java' 			AS tag_name UNION ALL
	SELECT '<c#>' 			AS tag_string, 'c_sharp' 		AS tag_name UNION ALL
	SELECT '<php>' 			AS tag_string, 'php' 			AS tag_name UNION ALL
	SELECT '<android>' 		AS tag_string, 'android' 		AS tag_name UNION ALL
	SELECT '<html>' 		AS tag_string, 'html' 			AS tag_name UNION ALL
	SELECT '<jquery>' 		AS tag_string, 'jquery' 		AS tag_name UNION ALL
	SELECT '<c++>' 			AS tag_string, 'c_plus_plus' 	AS tag_name UNION ALL
	SELECT '<css>' 			AS tag_string, 'css' 			AS tag_name UNION ALL
	SELECT '<ios>' 			AS tag_string, 'ios' 			AS tag_name UNION ALL
	SELECT '<sql>' 			AS tag_string, 'sql' 			AS tag_name UNION ALL
	SELECT '<mysql>' 		AS tag_string, 'mysql' 			AS tag_name UNION ALL
	SELECT '<r>' 			AS tag_string, 'r' 				AS tag_name UNION ALL
	SELECT '<reactjs>' 		AS tag_string, 'reactjs' 		AS tag_name UNION ALL
	SELECT '<node.js>' 		AS tag_string, 'node_js' 		AS tag_name UNION ALL
	SELECT '<arrays>' 		AS tag_string, 'arrays' 		AS tag_name UNION ALL
	SELECT '<c>' 			AS tag_string, 'c' 				AS tag_name UNION ALL
	SELECT '<asp.net>' 		AS tag_string, 'asp_net' 		AS tag_name UNION ALL
	SELECT '<json>' 		AS tag_string, 'json' 			AS tag_name
),
PrefilteredQuestions AS
(
	SELECT
		Id,
		Title,
		Body,
		CAST(Tags AS TEXT) AS tags_string,
		Score,
		YEAR(CreationDate) AS year
	FROM Posts
	WHERE
		PostTypeId = 1
		AND Score >= 50
		AND AnswerCount >= 5
		AND LEN(Tags) >= 1
		AND '2008-01-01' <= CreationDate
		AND CreationDate < '2024-01-01'
),
OneRowPerTagAndQuestion AS
(
	SELECT
		t.tag_name,
		q.Id,
		q.Title,
		q.Body,
		q.tags_string,
		q.year,
		ROW_NUMBER() OVER(PARTITION BY t.tag_name ORDER BY q.Score DESC, q.Id DESC) AS score_rank_in_tag
	FROM PrefilteredQuestions q
	INNER JOIN Tags t ON CHARINDEX(t.tag_string, q.tags_string) > 0
),
OneRowPerTagAndTop500Question AS
(
	SELECT
		tag_name,
		Id,
		Title,
		Body,
		tags_string,
		year,
		ROW_NUMBER() OVER(PARTITION BY Id ORDER BY tag_name) AS is_copy_to_keep_if_one
	FROM OneRowPerTagAndQuestion
	WHERE score_rank_in_tag <= 500
),
Top500Questions AS
(
	SELECT
		Id,
		Title,
		Body,
		tags_string,
		year
	FROM OneRowPerTagAndTop500Question
	WHERE is_copy_to_keep_if_one = 1
)
SELECT
	Id,
	Title,
	Body,
	tags_string,
	year,
	CASE WHEN CHARINDEX('<javascript>', tags_string) > 0 THEN 1 ELSE 0 END AS javascript,
	CASE WHEN CHARINDEX('<python>', tags_string) > 0 THEN 1 ELSE 0 END AS python,
	CASE WHEN CHARINDEX('<java>', tags_string) > 0 THEN 1 ELSE 0 END AS java,
	CASE WHEN CHARINDEX('<c#>', tags_string) > 0 THEN 1 ELSE 0 END AS c_sharp,
	CASE WHEN CHARINDEX('<php>', tags_string) > 0 THEN 1 ELSE 0 END AS php,
	CASE WHEN CHARINDEX('<android>', tags_string) > 0 THEN 1 ELSE 0 END AS android,
	CASE WHEN CHARINDEX('<html>', tags_string) > 0 THEN 1 ELSE 0 END AS html,
	CASE WHEN CHARINDEX('<jquery>', tags_string) > 0 THEN 1 ELSE 0 END AS jquery,
	CASE WHEN CHARINDEX('<c++>', tags_string) > 0 THEN 1 ELSE 0 END AS c_plus_plus,
	CASE WHEN CHARINDEX('<css>', tags_string) > 0 THEN 1 ELSE 0 END AS css,
	CASE WHEN CHARINDEX('<ios>', tags_string) > 0 THEN 1 ELSE 0 END AS ios,
	CASE WHEN CHARINDEX('<sql>', tags_string) > 0 THEN 1 ELSE 0 END AS sql,
	CASE WHEN CHARINDEX('<mysql>', tags_string) > 0 THEN 1 ELSE 0 END AS mysql,
	CASE WHEN CHARINDEX('<r>', tags_string) > 0 THEN 1 ELSE 0 END AS r,
	CASE WHEN CHARINDEX('<reactjs>', tags_string) > 0 THEN 1 ELSE 0 END AS reactjs,
	CASE WHEN CHARINDEX('<node.js>', tags_string) > 0 THEN 1 ELSE 0 END AS node_js,
	CASE WHEN CHARINDEX('<arrays>', tags_string) > 0 THEN 1 ELSE 0 END AS arrays,
	CASE WHEN CHARINDEX('<c>', tags_string) > 0 THEN 1 ELSE 0 END AS c,
	CASE WHEN CHARINDEX('<asp.net>', tags_string) > 0 THEN 1 ELSE 0 END AS asp_net,
	CASE WHEN CHARINDEX('<json>', tags_string) > 0 THEN 1 ELSE 0 END AS json
FROM Top500Questions
"""
```

# Feature engineering

```python
def clean_for_tf_idf(document):    
    TO_CONVERT = {
        'c#': 'csharp',
        'c++': 'cplusplus',
        'asp.net': 'aspdotnet',
        'node.js': 'nodejs'
    }
    
    def convert(string, to_convert=TO_CONVERT):
        for key, value in to_convert.items():
            string = string.replace(key, value)
        return string
    
    
    def remove_html_tags(string):
        html_tags = [
            '<blockquote>',
            '</blockquote>',
            '<br>',
            '</br>',
            '<code>',
            '</code>',
            '<em>',
            '</em>',
            '<p>',
            '</p>',
            '<pre>',
            '</pre>',
            '<strong>',
            '</strong>'
        ]
        for html_tag in html_tags:
            string = string.replace(html_tag, ' ')
        return string
    
    def keep_only_wanted_chars(string):
        string = re.sub(r'[^a-zA-Z\.,:;?!\']', ' ', string)
        return string
    
    TO_CONVERT_POST_TOKENIZATION = {
        'c': 'clanguage',
        'r': 'rlanguage'
    }
    
    def convert_post_tokenization(tokens, to_convert_post_tokenization=TO_CONVERT_POST_TOKENIZATION):
        for key, value in to_convert_post_tokenization.items():
            for i, token in enumerate(tokens):
                if token == key:
                    tokens[i] = value
        return tokens
    
    document = document.lower()
    document = convert(document)
    document = remove_html_tags(document)
    document = keep_only_wanted_chars(document)
    document = word_tokenize(document)
    document = convert_post_tokenization(document)
    document = ' '.join(document)
    return document
```

```python
# Creating and fitting title_vectorizer and body_vectorizer, then saving them with Pickle.
```

```python
title_vectorizer = TfidfVectorizer(max_features=2000)
cleaned_title = df['Title'].apply(clean_for_tf_idf)
title_vectorizer.fit(cleaned_title)
pickle.dump(title_vectorizer, open("title_vectorizer.sav", "wb"))

body_vectorizer = TfidfVectorizer(max_features=2000)
cleaned_body = df['Body'].apply(clean_for_tf_idf) 
body_vectorizer.fit(cleaned_body)
pickle.dump(body_vectorizer, open("body_vectorizer.sav", "wb"))
```

```python
def get_X_tf_idf(df, title_vectorizer, body_vectorizer):
    cleaned_title = df['Title'].apply(clean_for_tf_idf)
    title_tf_idf_features = title_vectorizer.transform(cleaned_title)
    
    title_tf_idf_features_df = pd.DataFrame(
        data=title_tf_idf_features.toarray(),
        columns=['title_' + feature_name for feature_name in title_vectorizer.get_feature_names_out()]
    )
    
    cleaned_body = df['Body'].apply(clean_for_tf_idf)
    body_tf_idf_features = body_vectorizer.transform(cleaned_body)

    body_tf_idf_features_df = pd.DataFrame(
        data=body_tf_idf_features.toarray(),
        columns=['body_' + feature_name for feature_name in body_vectorizer.get_feature_names_out()]
    )

    X_tf_idf = pd.concat(
        [
            title_tf_idf_features_df,
            body_tf_idf_features_df
        ],
        axis=1
    )
    
    return X_tf_idf
```

```python
def get_X_word2vec(df):
    
    html_tags = [
        '<blockquote>',
        '</blockquote>',
        '<br>',
        '</br>',
        '<code>',
        '</code>',
        '<em>',
        '</em>',
        '<p>',
        '</p>',
        '<pre>',
        '</pre>',
        '<strong>',
        '</strong>'
    ]
    
    def remove_substrings(sentence, substrings=html_tags):
        for substring in substrings:
            sentence = sentence.replace(substring, ' ')
        return sentence
    
    def remove_non_letters(sentence):
        sentence = re.sub(r'[^a-zA-Z]', ' ', sentence)
        return sentence
    
    def remove_html_tags_and_non_letters_and_split(sentence):
        sentence = remove_substrings(sentence)
        sentence = remove_non_letters(sentence)
        split_words = sentence.split()
        return split_words

    def convert_split_words_to_vectors(split_words, vector_dict):
        word_vectors = []
        for word in split_words:
            try:
                word_vectors.append(vector_dict.get_vector(word))
            except:
                continue
        
        if not word_vectors:
            word_vectors = np.zeros((1, vector_dict.vector_size))
        
        return np.array(word_vectors)
    
    def convert_word_vectors_to_sentence_vector(word_vectors):
        return word_vectors.mean(axis=0)
    
    def convert_split_words_to_sentence_vector(split_words):
        word_vectors = convert_split_words_to_vectors(split_words, word2vec_vectors)
        sentence_vector = convert_word_vectors_to_sentence_vector(word_vectors)
        return sentence_vector
    
    word2vec_vectors = gensim.downloader.load('word2vec-google-news-300')
    
    split_text = pd.Series(df['Title'].apply(remove_html_tags_and_non_letters_and_split))

    sentence_vectors = split_text.apply(convert_split_words_to_sentence_vector)
    
    X_word2vec = pd.DataFrame(np.array(sentence_vectors.tolist()))
    
    return X_word2vec
```

```python
def get_X_bert(df):
    
    def remove_html_tags(string):
        html_tags = [
            '<blockquote>',
            '</blockquote>',
            '<br>',
            '</br>',
            '<code>',
            '</code>',
            '<em>',
            '</em>',
            '<p>',
            '</p>',
            '<pre>',
            '</pre>',
            '<strong>',
            '</strong>'
        ]
        for html_tag in html_tags:
            string = string.replace(html_tag, ' ')
        return string
    
    def keep_only_wanted_chars(string):
        string = re.sub(r'[^a-zA-Z\.,:;?!\']', ' ', string)
        return string
    
    def clean(document):
        document = remove_html_tags(document)
        return document
    
    cleaned_title = df['Title'].apply(clean)
    cleaned_body = df['Body'].apply(clean)

    vectorizer = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L3-v2')

    # Feature creation
    title_embeddings = vectorizer.encode(cleaned_title)
    body_embeddings = vectorizer.encode(cleaned_body)
    
    X_bert = pd.DataFrame(np.hstack([title_embeddings, body_embeddings]))
    
    return X_bert
```

```python
def get_X_use(df):
    
    def remove_html_tags(string):
        html_tags = [
            '<blockquote>',
            '</blockquote>',
            '<br>',
            '</br>',
            '<code>',
            '</code>',
            '<em>',
            '</em>',
            '<p>',
            '</p>',
            '<pre>',
            '</pre>',
            '<strong>',
            '</strong>'
        ]
        for html_tag in html_tags:
            string = string.replace(html_tag, ' ')
        return string
    
    def keep_only_wanted_chars(string):
        string = re.sub(r'[^a-zA-Z\.,:;?!\']', ' ', string)
        return string
    
    def clean(document):
        document = remove_html_tags(document)
        return document
    
    cleaned_title = df['Title'].apply(clean)
    cleaned_body = df['Body'].apply(clean)

    vectorizer = SentenceTransformer('sentence-transformers/use-cmlm-multilingual')

    # Feature creation
    title_embeddings = vectorizer.encode(cleaned_title)
    body_embeddings = vectorizer.encode(cleaned_body)
    
    X_use = pd.DataFrame(np.hstack([title_embeddings, body_embeddings]))
    return X_use
```
